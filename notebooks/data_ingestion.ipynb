```python
import os
import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine

# 1. Asegurarnos que estamos apuntando a la raíz del proyecto para encontrar la carpeta 'data' y el '.env'
if os.path.basename(os.getcwd()) == 'notebooks':
    os.chdir('..')

# 2. Cargar nuestras llaves guardadas en el archivo .env
load_dotenv()
database_url = os.getenv("DATABASE_URL")

# 3. Leer el dataset crudo
ruta_csv = 'data/raw/steam-games.csv'
print(f"⏳ Leyendo archivo f{ruta_csv}...")
df_games = pd.read_csv(ruta_csv)

# 4. Limpieza rápida de nombres de columnas (quitar espacios ayuda mucho en bases de datos)
df_games.columns = df_games.columns.str.lower().str.replace(' ', '_')

# 5. Conectarse a Supabase de forma directa
print("🚀 Conectando a Supabase...")
engine = create_engine(database_url)

# 6. Crear la tabla y subir todos los datos en un solo movimiento
print("☁️ Subiendo catálogo de juegos a la nube (esto puede tardar un poco dependiendo del tamaño)...")
df_games.to_sql('steam_games', engine, if_exists='replace', index=False)

print("✅ ¡Éxito! Tu tabla de catálogo ha sido creada en la nube de Supabase. Todo tu grupo ya puede verla.")
```
