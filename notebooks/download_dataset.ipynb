```python
import kagglehub


path = kagglehub.dataset_download("antonkozyriev/game-recommendations-on-steam")

print("Path to dataset files:", path)
```

```python
import shutil
import os

if os.path.basename(os.getcwd()) == 'notebooks':
    os.chdir('..')

archivos_clave = ["games.csv", "recommendations.csv", "users.csv", "games_metadata.json"]
destino = "data/raw/"
os.makedirs(destino, exist_ok=True)
print("\n📦 Moving files to -> data/raw/ ...")
for archivo in archivos_clave:
    origen = os.path.join(path, archivo)
    if os.path.exists(origen):
        shutil.copy(origen, os.path.join(destino, archivo))
        print(f"   -> Now moving: {archivo}")
    else:
        print(f"   ⚠️ Missing: {archivo}")
print("\n🚀 ¡READY!.")
```
