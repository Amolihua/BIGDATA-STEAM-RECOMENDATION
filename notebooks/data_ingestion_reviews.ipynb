```python
import pandas as pd
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine
```

```python
if os.path.basename(os.getcwd()) == 'notebooks':
    os.chdir('..')

load_dotenv()
engine = create_engine(os.getenv("DATABASE_URL"))

print("Descomprimiendo el Parquet local...")
df_reviews = pd.read_parquet('data/processed/steam_interactions_v1.parquet')

print(f"Iniciando subida de {len(df_reviews)} reseñas a Supabase...")
df_reviews.to_sql('steam_reviews', engine, if_exists='replace', index=False, chunksize=5000)

print("Proceso termindado")
```
