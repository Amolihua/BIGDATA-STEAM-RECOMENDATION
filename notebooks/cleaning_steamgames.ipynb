```python
import pandas as pd
```

```python
df_games = pd.read_csv('../data/raw/steam-games.csv')
```

```python
df_games.head()
```

```python
df_games.columns = df_games.columns.str.lower().str.replace(' ', '_')

df_clean = df_games.dropna(subset=['app_id', 'overall_review_%']).copy()

df_clean['app_id'] = df_clean['app_id'].astype(int)
```

```python

ruta_guardado = '../data/processed/steam_games_cleaned.csv'
df_clean.to_csv(ruta_guardado, index=False)
print(f"✅ Catálogo limpio guardado exitosamente en: {ruta_guardado}")
```
